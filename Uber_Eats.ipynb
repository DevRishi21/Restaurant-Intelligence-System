## Data Pre-Processing

```python
# remove duplicated rows
restaurants_df = restaurants_df.drop_duplicates()
```

```python
# normalize rate column
restaurants_df['rate'] = (restaurants_df['rate'].str.replace("/5","").str.strip().replace("NEW", np.nan))
```

```python
# convert rate column to numeric
restaurants_df['rate'] = pd.to_numeric(restaurants_df["rate"])
```

```python
# normalize approx_cost column
restaurants_df["approx_cost(for two people)"] = (restaurants_df["approx_cost(for two people)"].str.replace(",","").str.strip())
```

```python
# convert approx_cost column to numeric
restaurants_df["approx_cost(for two people)"] = pd.to_numeric(restaurants_df["approx_cost(for two people)"])
```

## Feature Engineering

```python
# analyze cost distribution to create pricing buckets
restaurants_df["approx_cost(for two people)"] .describe()
```

```python
# function to map pricing segment
def map_pricing_segment(cost):
    if(pd.isna(cost)):
        return "Unknown"
    elif cost <= 400:
        return "Budget"
    elif cost <= 600:
        return "Mid Range"
    elif cost <= 1000:
        return "Premium"
    else:
        return "Elite"

# function to map ratings category
def map_ratings(rating):
    if pd.isna(rating):
        return "Unrated"
    elif rating < 3:
        return "Poor"
    elif rating >= 3 and rating <= 4:
        return "Good"
    else:
        return "Excellent"
```

```python
# create new column and map segments
restaurants_df["pricing_segment"] = restaurants_df["approx_cost(for two people)"].apply(map_pricing_segment)
```

```python
# similarly create new column bucket for ratings
restaurants_df['ratings_category'] = restaurants_df['rate'].apply(map_ratings)
```

```python
restaurants_df.isnull().sum()
```

```python
# renaming columns
restaurants_df = restaurants_df.rename(
    columns = {
        'approx_cost(for two people)': 'approx_cost_for_two',
        'listed_in(type)': 'listed_in_type',
        'listed_in(city)': 'listed_in_city'
    }
)
```

## Resolve Business Queries

### Which Bangalore locations have the highest average restaurant ratings?

```python
# Changing float type format globally
pd.options.display.float_format = '{:.1f}'.format
```

```python
q1 = """
SELECT location, AVG(rate) as avg_ratings, ratings_category
FROM restaurants
WHERE rate is NOT NULL
GROUP BY location, ratings_category
ORDER BY avg_ratings DESC
"""

q1_result = pd.read_sql(q1, engine)
q1_result
```

### Which locations are over-saturated with restaurants?

```python
q2 = """
SELECT location, COUNT(location) as total_restaurants
FROM restaurants
GROUP BY location
ORDER BY total_restaurants DESC
"""

q2_result = pd.read_sql(q2, engine)
q2_result
```

### Does online ordering improve restaurant ratings?

```python
q3 = """
SELECT online_order, ROUND(AVG(rate), 2) AS avg_rating, COUNT(*) AS restaurant_count
FROM restaurants
WHERE rate IS NOT NULL
GROUP BY online_order;
"""

q3_result = pd.read_sql(q3, engine)
q3_result
```

### Does table booking correlate with higher customer ratings?

```python
q4 = """
SELECT book_table as table_booking, ROUND(AVG(rate), 2) AS avg_rating, COUNT(*) AS restaurant_count
FROM restaurants
WHERE rate IS NOT NULL
GROUP BY table_booking;
"""

q4_result = pd.read_sql(q4, engine)
q4_result
```

### What price range delivers the best customer satisfaction?

```python
q5 = """
SELECT AVG(approx_cost_for_two) as avg_cost_for_two, rate as ratings, pricing_segment
FROM restaurants
WHERE rate is NOT NULL
GROUP BY ratings, pricing_segment
ORDER BY ratings DESC
"""

q5_result = pd.read_sql(q5, engine)
q5_result
```

### How do low, mid, and premium-priced restaurants perform in terms of ratings?

```python
q6 = """
SELECT
    pricing_segment,
    ROUND(AVG(rate), 2) AS avg_rating,
    COUNT(*) AS restaurant_count
FROM restaurants
WHERE rate IS NOT NULL
GROUP BY pricing_segment
ORDER BY avg_rating DESC;
"""

q6_result = pd.read_sql(q6, engine)
q6_result
```

### Which cuisines are most common in Bangalore?

```python
q7 = """
SELECT
    cuisines,
    COUNT(*) AS restaurant_count
FROM restaurants
GROUP BY cuisines
ORDER BY restaurant_count DESC
LIMIT 10;
"""

q7_result = pd.read_sql(q7, engine)
q7_result
```

### Which cuisines receive the highest average ratings?

```python
q8 = """
SELECT
    cuisines,
    ROUND(AVG(rate), 2) AS avg_rating,
    COUNT(*) AS restaurant_count
FROM restaurants
WHERE rate IS NOT NULL
GROUP BY cuisines
HAVING COUNT(*) >= 20
ORDER BY avg_rating DESC
LIMIT 10;
"""

q8_result = pd.read_sql(q8, engine)
q8_result
```

### Which cuisines perform well despite having fewer restaurants?

```python
q9 = """
SELECT
    cuisines,
    COUNT(*) AS restaurant_count,
    ROUND(AVG(rate), 2) AS avg_rating
FROM restaurants
WHERE rate IS NOT NULL
GROUP BY cuisines
HAVING COUNT(*) BETWEEN 20 AND 100
ORDER BY avg_rating DESC;
"""

q9_result = pd.read_sql(q9, engine)
q9_result
```

### What is the relationship between restaurant cost and rating?

```python
q10 = """
SELECT
    pricing_segment,
    ROUND(AVG(rate), 2) AS avg_rating,
    ROUND(AVG(votes), 0) AS avg_votes,
    COUNT(*) AS restaurant_count
FROM restaurants
WHERE rate IS NOT NULL
GROUP BY pricing_segment
ORDER BY AVG(approx_cost_for_two);
"""

q10_result = pd.read_sql(q10, engine)
q10_result
```

## Business queries for Orders

### Which restaurants generate the highest order revenue?

```python
q11 = """
SELECT
    restaurant_name,
    ROUND(SUM(order_value), 2) AS total_revenue,
    COUNT(*) AS total_orders
FROM orders
GROUP BY restaurant_name
ORDER BY total_revenue DESC
LIMIT 10;
"""

q11_result = pd.read_sql(q11, engine)
q11_result
```

### Which months receive the highest number of orders?

```python
q12 = """
SELECT
    TO_CHAR(order_date, 'MONTH') AS month,
    COUNT(*) AS total_orders
FROM orders
GROUP BY month
ORDER BY total_orders DESC;
"""

q12_result = pd.read_sql(q12, engine)
q12_result
```

### Do discounts increase customer spending?

```python
q13 = """
SELECT
    discount_used,
    ROUND(AVG(order_value), 2) AS avg_order_value,
    COUNT(*) AS total_orders
FROM orders
GROUP BY discount_used;
"""

q13_result = pd.read_sql(q13, engine)
q13_result
```

### Which payment methods are most preferred?

```python
q14 = """
SELECT
    payment_method,
    COUNT(*) AS total_orders,
    ROUND(AVG(order_value), 2) AS avg_order_value
FROM orders
GROUP BY payment_method
ORDER BY total_orders DESC;
"""

q14_result = pd.read_sql(q14, engine)
q14_result
```

### Which restaurants have the highest average order value?

```python
q15 = """
SELECT
    restaurant_name,
    ROUND(AVG(order_value), 2) AS avg_order_value,
    COUNT(*) AS total_orders
FROM orders
GROUP BY restaurant_name
ORDER BY avg_order_value DESC
LIMIT 10;
"""

q15_result = pd.read_sql(q15, engine)
q15_result
```
